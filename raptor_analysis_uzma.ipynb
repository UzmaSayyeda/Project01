### Q4 minutes played vs pace impact in relevance to raptor total



```python
pace_min_played=clean_df.groupby(["player_name"])[["mp","pace_impact","raptor_total"]].sum()
pace_min_played.head()
```

```python
pace_min_played=pace_min_played.reset_index()
pace_min_played
```

```python
pace=pace_min_played.groupby(["player_name"])[["pace_impact","mp","raptor_total"]].sum().sort_values(by="pace_impact",ascending=False)
pace=pace.reset_index()
pace=pace.head()
pace
```

```python
rap_mp=pace_min_played.groupby("player_name")[["mp","pace_impact","raptor_total"]].sum().sort_values(by="mp",ascending=False)

rap_mp=rap_mp.reset_index()
rap_mp=rap_mp.head()
rap_mp
```

```python
merged_pace_mp=pd.concat([rap_mp,pace],join="inner",ignore_index=True)
merged_pace_mp
```

```python
merged_pace_mp.sort_values(by="pace_impact",ascending=False)
```

```python
players=merged_pace_mp["player_name"]
mp_02=merged_pace_mp["mp"]
pace_02=merged_pace_mp["pace_impact"]
rt=merged_pace_mp["raptor_total"]
```

```python
plt.scatter(mp_02,pace_02)
plt.xlabel("Player Name")
plt.ylabel("Pace Impact")
plt.title("Minutes Played vs Pace Impact")
plt.xticks(mp_02,players,rotation=90)
plt.show()
```

```python
plt.bar(mp_02,pace_02)
plt.xticks(mp_02,rotation=90)
plt.xlabel("Minutes Played")
plt.ylabel("Pace Impact")
plt.title("Minutes Played vs Pace Impact")
plt.show()
```

```python
plt.barh(players,mp_02)
plt.xlabel("Minutes Played")
plt.ylabel("Player Names")
plt.title("Players vs. Minutes Played")
plt.xticks()

plt.show()
```

```python
plt.bar(players,pace_02)
plt.xlabel("Player Names")
plt.ylabel("Pace Impact")
plt.title("Players vs. Pace Impact")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(players,rt)
plt.title("Players vs RAPTOR total")
plt.xlabel("Player Names")
plt.ylabel("RAPTOR totals")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(players,rt)
plt.bar(players,pace_02)
plt.title("Players vs RAPTOR total")
plt.xlabel("Player Names")
plt.ylabel("RAPTOR totals")
plt.xticks(rotation=90)
plt.show()
```
