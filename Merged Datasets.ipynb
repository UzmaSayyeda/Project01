# NBA RAPTOR Stat Exploratory Comparison - Q1: Posessions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

latest_df = pd.read_csv("CSVs/latest_RAPTOR_by_team.csv")
historical_df = pd.read_csv("CSVs/historical_RAPTOR_by_team.csv")
```

```python
latest_df.head()
```

```python
historical_df.head()
```

```python
# merging both datasets

merged_df = pd.merge(historical_df, latest_df, how="outer")
merged_df
```

```python
merged_df["season"].unique()
```

```python
merged_df.isnull().sum()
```

```python
merged_df.fillna(0, inplace=True)
```

```python
# showcasing that there is no more NaN/Null
clean_merged_df = merged_df
clean_merged_df.isnull().sum()
```

```python
clean_merged_df
```

## Do the amount of possessions per player greatly impact that player’s overall RAPTOR ratings? Why or why not?

```python
# creating a full list of players from both sets of data based on possessions and total RAPTOR scores.

possession = clean_merged_df.groupby("player_name")["poss"].sum()
raptor_total = clean_merged_df.groupby("player_name")["raptor_total"].sum()

possession_df = pd.DataFrame(possession)
possession_df.sort_values(by="poss", ascending=False, inplace=True)

raptor_total_df = pd.DataFrame(raptor_total)
raptor_total_df.sort_values(by="raptor_total", ascending=False, inplace=True)
```

```python
possession_df
```

```python
raptor_total
```

```python
# defining top 10 players by possession/total RAPTOR score

possession_top = possession_df.head(10)
raptor_total_top = raptor_total_df.head(10)
```

```python
# graphing top 10 players by possession

ax = possession_top.plot(kind="barh", figsize=(20,15))
ax.bar_label(ax.containers[0])
plt.title("Top 10 Players by Possession, 1977-2023", weight="bold", fontsize=20)
plt.xlabel("Possessions", weight="bold", fontsize=16)
plt.ylabel("Player Name", weight="bold", fontsize=16)
plt.gca().invert_yaxis()
    
plt.show()
```

```python
# graphing top 10 players by possession

ax = raptor_total_top.plot(kind="barh", figsize=(20,15))
ax.bar_label(ax.containers[0])
plt.title("Top 10 Players by RAPTOR Total, 1977-2023", weight="bold", fontsize=20)
plt.xlabel("RAPTOR Score", weight="bold", fontsize=16)
plt.ylabel("Player Name", weight="bold", fontsize=16)
plt.gca().invert_yaxis()
    
plt.show()
```

```python
# combining poss/total RAPTOR score
combined_df = pd.DataFrame(clean_merged_df.groupby("player_name")["raptor_total","poss"].sum())
```

```python
top_10_poss = combined_df.sort_values(by="poss", ascending=False).head(10)
top_10_poss.head(10)
```

```python
top_10_raptor = combined_df.sort_values(by="raptor_total", ascending=False).head(10)
top_10_raptor
```

```python
ax = top_10_poss.plot(kind="barh", figsize=(20,15))
ax.bar_label(ax.containers[1])
plt.title("Top 10 Players by RAPTOR Total, 1977-2023", weight="bold", fontsize=20)
plt.xlabel("Player Name", weight="bold", fontsize=16)
plt.ylabel("RAPTOR Score", weight="bold", fontsize=16)
#plt.gca().invert_yaxis()
    
plt.show()
```

```python
top_10_poss['player_name'] = top_10_poss.groupby('poss', sort=False).ngroup() + 1
#top_10_poss['player_name'].replace({1: 'x'})
top_10_poss
```

```python
x = top_10_poss["raptor_total"]
y = top_10_poss["poss"]

# plot x and y on scatter plot
plt.scatter(x, y)
# add axes labels
plt.xlabel("Total RAPTOR Score")
plt.ylabel("Possessions")
# add labels to all points
for (xi, yi) in zip(x, y):
    plt.text(xi, yi, yi, va='bottom', ha='center')
```

```python
# Build scatter plot for RAPTOR vs. possession
plt.plot(top_10_poss["raptor_total"], 
         top_10_poss["poss"],
         linestyle="none", 
         marker="o", 
         markeredgecolor="black", 
         markeredgewidth=0.75)

# Incorporate the other graph properties
plt.xlabel("Total RAPTOR Score")
plt.ylabel("Possessions")
plt.title("Top 10 Players /w Highest Possessions vs. RAPTOR Scores")
plt.grid()

# Save the figure
# plt.savefig("output_data/Fig1.png")

# Show plot
plt.show()
```
