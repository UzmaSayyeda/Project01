```python
# categories that describe top players
raptor_war = stats_2023.groupby("player_name")["war_total"].sum()
raptor_war_df = pd.DataFrame(raptor_war)
raptor_war_df = raptor_war_df.sort_values(by="war_total",ascending=False)

raptor_total = stats_2023.groupby("player_name")["raptor_total"].sum()
raptor_total_df = pd.DataFrame(raptor_total)
raptor_total_df = raptor_total_df.sort_values(by="raptor_total",ascending=False)


raptor_offense = stats_2023.groupby("player_name")["raptor_offense"].sum()
raptor_offense_df = pd.DataFrame(raptor_offense)
raptor_offense_df = raptor_offense_df.sort_values(by="raptor_offense",ascending=False)

raptor_defense = stats_2023.groupby("player_name")["raptor_defense"].sum()
raptor_defense_df = pd.DataFrame(raptor_defense)
raptor_defense_df = raptor_defense_df.sort_values(by="raptor_defense",ascending=False)

# add mp to categories to give correct top players
```
